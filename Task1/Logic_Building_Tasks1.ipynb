### Task 2: Pass / Fail Analyzer

```python
# Problem 2: Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

# Initialize counters
pass_count = 0
fail_count = 0

# Loop through each student's mark
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print the final analysis
print("--- Academic Evaluation Results ---")
print(f"Total students passed: {pass_count}")
print(f"Total students failed: {fail_count}")
```

### Task 3: Simple Data Cleaner

```python
# Problem 3: Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

# Clean data: strip spaces and convert to lowercase for each name in the list
cleaned_names = [name.strip().lower() for name in names]

# Print the standardized list
print("Cleaned names list:", cleaned_names)
```

### Task 4: Message Length Analyzer

```python
# Problem 4: Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

print("--- Message Analysis ---")

# Loop through each message to analyze its length
for msg in messages:
    msg_length = len(msg)
    
    # Print length and apply flag if it exceeds 10 characters
    if msg_length > 10:
        print(f"Message: '{msg}' | Length: {msg_length} -> [FLAGGED: > 10 chars]")
    else:
        print(f"Message: '{msg}' | Length: {msg_length}")
```

### Task5: Error Message Detector

```python
# Problem 5: Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Use the list count() method to find exact matches of "ERROR"
error_count = logs.count("ERROR")

# Print the final count
print(f"Total system errors detected: {error_count}")
```
