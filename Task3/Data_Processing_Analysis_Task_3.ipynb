# --- Problem 1: Employee Performance Bonus Eligibility ---

```python
print("\n--- Problem 1: Employee Performance Bonus ---")
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# 1. Find the maximum score from the dictionary values
max_score = max(employees.values())

# 2. Identify all employees who achieved this maximum score
top_performers = [name for name, score in employees.items() if score == max_score]

# 3. Print the result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")
```

# --- Problem 2: Search Query Keyword Analysis ---

```python
print("\n--- Problem 2: Search Query Analysis ---")
search_query = "Buy mobile phone buy phone online"

# 1. Convert to lowercase and split into words
words = search_query.lower().split()

# 2. Count frequency of each word
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1

# 3. Filter for words that appear more than once
duplicates = {word: count for word, count in word_count.items() if count > 1}
print(duplicates)
```

# --- Problem 3: Sensor Data Validation ---

```python
print("\n--- Problem 3: Sensor Data Validation ---")
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []

# 1. Loop through the list with index using enumerate()
for hour, reading in enumerate(sensor_readings):
    # 2. Check if reading is even (valid)
    if reading % 2 == 0:
        valid_readings.append((hour, reading))

print("Valid Sensor Readings (Hour, Value):", valid_readings)
```

# --- Problem 4: Email Domain Usage Analysis ---

```python
print("\n--- Problem 4: Email Domain Analysis ---")
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_counts = {}

# 1. Extract domains and count them
for email in emails:
    domain = email.split('@')[1]  # Split by @ and take the second part
    domain_counts[domain] = domain_counts.get(domain, 0) + 1

# 2. Calculate percentage and print
total_users = len(emails)
for domain, count in domain_counts.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage}%")
```

# --- Problem 5: Sales Spike Detection ---

```python
print("\n--- Problem 5: Sales Spike Detection ---")
sales = [1200, 1500, 900, 2200, 1400, 3000]

# 1. Calculate average sales
average_sales = sum(sales) / len(sales)
threshold = average_sales * 1.30  # 30% above average

# 2. Detect spikes
for day, sale in enumerate(sales):
    if sale > threshold:
        # Adding 1 to day index because days usually start at 1, not 0
        print(f"Day {day + 1}: {sale}")
```

# --- Problem 6: Duplicate User ID Detection ---

```python
print("\n--- Problem 6: Duplicate User ID Detection ---")
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
id_counts = {}

# 1. Count occurrences of each User ID
for user in user_ids:
    id_counts[user] = id_counts.get(user, 0) + 1

# 2. Display only duplicates
for user, count in id_counts.items():
    if count > 1:
        print(f"{user} {count} times")
```
