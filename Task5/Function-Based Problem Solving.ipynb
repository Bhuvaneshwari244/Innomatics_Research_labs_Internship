# Problem 1: Social Media - Post Engagement Analyzer

```python
def analyze_engagement(likes_list):
    total_likes = 0
    # Use loop for calculations [cite: 40]
    for like in likes_list:
        total_likes += like
    
    # Conditional logic for post status [cite: 33, 34]
    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"
    
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

# Example usage
analyze_engagement([500, 200, 350])
```

# Problem 3: Agriculture - Rainfall Adequacy Checker

```python
def check_rainfall_adequacy(rainfall_data, required_level):
    # Calculate average rainfall using loop and math [cite: 68, 76]
    average_rainfall = sum(rainfall_data) / len(rainfall_data)
    
    # Determine adequacy [cite: 69, 70]
    status = "Adequate Rainfall" if average_rainfall >= required_level else "Inadequate Rainfall"
    
    print(f"Average Rainfall: {average_rainfall}")
    print(f"Rainfall Status: {status}")

check_rainfall_adequacy([70, 80, 66], 72)
```

# Problem 4: Social Media - Duplicate Account Detection

```python
def detect_duplicates(usernames):
    # Use set data structure to find unique names [cite: 92]
    if len(usernames) != len(set(usernames)):
        found = "Yes"
    else:
        found = "No"
    
    print(f"Duplicate Accounts Found: {found}")

detect_duplicates(["user1", "user2", "user1"])
```

# Problem 6: Agriculture - Premium Crop Price Filter

```python
def filter_premium_crops(prices):
    # Use list filtering logic for prices > 2000 [cite: 117, 124]
    premium_crops = [price for price in prices if price > 2000]
    
    print(f"Premium Crops: {premium_crops}")

filter_premium_crops([1500, 2500, 1800, 3200])
```

# Problem 7: System Monitoring - Application Health Checker

```python
def evaluate_app_health(error_count):
    # Conditional logic for system status [cite: 132, 133, 134]
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")

evaluate_app_health(7)
```

# Problem 9: E-Learning - Student Attendance Eligibility System

```python
def check_exam_eligibility(attendance_list):
    # Calculate percentage based on attended sessions [cite: 167, 176]
    attendance_pct = (sum(attendance_list) / len(attendance_list)) * 100
    
    # 75% threshold for eligibility [cite: 168, 169]
    eligibility = "Eligible" if attendance_pct >= 75 else "Not Eligible"
    
    print(f"Attendance Percentage: {attendance_pct}")
    print(f"Exam Eligibility: {eligibility}")

# 1 for present, 0 for absent
check_exam_eligibility([1, 1, 1, 1, 0])
```

# Problem 10: Smart Electricity Bill Analyzer

```python
def analyze_electricity_bill(units):
    bill = 0
    # Billing Rules: Progressive slabs [cite: 184, 185, 186, 187]
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Usage Classification [cite: 188, 190]
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
        
    return bill, usage # [cite: 197]

# Implementation and printing
units_consumed = 250
bill_amount, usage_status = analyze_electricity_bill(units_consumed)
print(f"Units Consumed: {units_consumed}")
print(f"Total Bill: ₹{bill_amount}")
print(f"Usage Level: {usage_status}")
```
