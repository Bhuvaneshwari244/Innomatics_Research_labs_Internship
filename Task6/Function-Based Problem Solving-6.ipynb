# Problem Statement 1: Smart Parking Lot Management System

```python
def manage_parking_lot(capacity, logs):
    parked_vehicles = 0
    
    # Loop for counting entries and exits
    for log in logs:
        if log == "IN":
            parked_vehicles += 1
        elif log == "OUT":
            parked_vehicles -= 1
            
    # Conditional alert for parking status
    if parked_vehicles >= capacity:
        status = "Full"
    else:
        status = "Available"
        
    print(f"Currently Parked Vehicles: {parked_vehicles}")
    print(f"Parking Status: {status}")

# Sample Input Execution
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
manage_parking_lot(50, vehicle_logs)
```

# Problem Statement 2: Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):
    # Base calculation: Assuming 5 minutes per kilometer
    base_time = distance * 5
    
    # Mathematical adjustments for traffic conditions
    if traffic.lower() == "high":
        base_time += 10
    elif traffic.lower() == "medium":
        base_time += 5
        
    # Mathematical adjustments for weather conditions
    if weather.lower() == "rainy":
        base_time += 5
        
    print(f"Estimated Delivery Time: {base_time} minutes")

# Sample Input Execution
estimate_delivery_time(8, "High", "Rainy")
```

# Problem Statement 3: Movie Theatre Seat Occupancy Analyzer

```python
def analyze_seat_occupancy(total_seats, booked_seats):
    # Calculate total booked seats from the list
    booked_count = len(booked_seats)
    
    # Percentage calculation
    occupancy_pct = (booked_count / total_seats) * 100
    
    # Determine if show is Housefull or Almost Full
    if occupancy_pct >= 100:
        status = "Housefull"
    elif occupancy_pct >= 75:
        status = "Almost Full"
    else:
        status = "Available"
        
    print(f"Occupancy: {int(occupancy_pct)}%")
    print(f"Show Status: {status}")

# Sample Input Execution
# Creating a list of 150 entries to match the sample input requirement
booked_seats_list = [1] * 150 
analyze_seat_occupancy(200, booked_seats_list)
```

# Problem Statement 4: Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    # Average calculation
    avg_cpu = sum(cpu_readings) / len(cpu_readings)
    
    # Classify load based on rules
    if avg_cpu > 80:
        status = "Critical"
    elif avg_cpu >= 50:
        status = "Warning"
    else:
        status = "Normal"
        
    print(f"Average CPU Load: {int(avg_cpu)}%")
    print(f"Server Status: {status}")

# Sample Input Execution
cpu_data = [45, 60, 70, 85, 90]
classify_server_load(cpu_data)
```

# Problem Statement 5: Smart Classroom Resource Usage Monitor

```python
def monitor_resource_usage(usage_dict):
    overused = []
    
    # Iterate through dictionary to find overused resources (assuming > 8 hours is overuse)
    for resource, hours in usage_dict.items():
        if hours > 8:
            overused.append(resource)
            
    # Conditional checks for Energy Alert
    energy_alert = "Yes" if overused else "No"
    
    if overused:
        print(f"Overused Resources: {', '.join(overused)}")
    print(f"Energy Alert: {energy_alert}")

# Sample Input Execution
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
monitor_resource_usage(resources)
```

# Problem Statement 6: Online Event Registration Capacity Controller

```python
def manage_event_registrations(capacity, total_registrations):
    confirmed = 0
    waitlisted = 0
    
    # Loop to track registrations and prevent overbooking
    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
            
    # Determine registration status
    status = "Closed" if confirmed >= capacity else "Open"
    
    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")

# Sample Input Execution
manage_event_registrations(100, 105)
```
