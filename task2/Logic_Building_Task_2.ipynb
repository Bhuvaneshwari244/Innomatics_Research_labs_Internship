# --- Problem 1: Unique Words in a Sentence ---

```python
print("\n--- Problem 1: Unique Words ---")
sentence = "python is easy and python is powerful"
words = sentence.split()             # Split sentence into a list of words
unique_words = set(words)            # Convert to set to remove duplicates
print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

# --- Problem 2: Highest Salary from Employee Data ---

```python
print("\n--- Problem 2: Highest Salary ---")
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
# Find the key (employee) with the maximum value (salary)
highest_earner = max(employees, key=employees.get)
print(f"Highest Salary: {highest_earner} - {employees[highest_earner]}")
```

# --- Problem 4: Count Products Above a Price Threshold ---

```python
print("\n--- Problem 4: Product Price Threshold ---")
prices = [450, 1200, 899, 1500, 300]
count_above_1000 = 0
for price in prices:
    if price > 1000:
        count_above_1000 += 1
print(f"Products above 1000: {count_above_1000}")
```

# --- Problem 5: Calculate Attendance Percentage ---

```python
print("\n--- Problem 5: Attendance Percentage ---")
attendance = ["P", "P", "A", "P", "P"]
total_days = len(attendance)
days_present = attendance.count("P")
percentage = (days_present / total_days) * 100
print(f"Attendance Percentage: {percentage}")
```

# --- Problem 7: Count Character Frequency ---

```python
print("\n--- Problem 7: Character Frequency ---")
input_str = "python"
char_count = {}
for char in input_str:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1
print(char_count)
```

# --- Problem 9: Check if a Key Exists in a Dictionary ---

```python
print("\n--- Problem 9: Key Existence Check ---")
employee_dict = {"Ravi": 75000, "Anita": 68000}
key_to_check = "Ravi"
if key_to_check in employee_dict:
    print("Employee exists")
else:
    print("Employee not found")
```
