# --- Problem 1: Mobile Recharge Validation System ---

```python
def process_recharge():
    valid_plans = [199, 299, 399, 599] # Valid plans according to rules [cite: 343]
    
    while True: # Use while loop for retry [cite: 350]
        try:
            amount = int(input("Enter recharge amount (₹): "))
            if amount < 50: # Recharge amount must be >= 50 [cite: 342]
                print("Invalid: Recharge amount must be ₹50 or more. Please try again.")
                continue
                
            if amount in valid_plans:
                print(f"Success: Recharge of ₹{amount} is successful!")
                break
            else:
                print(f"Invalid Plan. Valid plans are {valid_plans}. Please try again.")
        except ValueError:
            print("Please enter a valid number.")

print("\n--- Problem 1 Output ---")
process_recharge() # Uncomment this line to run the interactive prompt
```

# --- Problem 2: Inventory Reorder Alert System ---

```python
def check_inventory(inventory_dict):
    print("\n--- Problem 2 Output ---")
    # Use for loop on dictionary [cite: 363]
    for product, stock in inventory_dict.items():
        if stock < 15: # Print Reorder Alert if stock < 15 [cite: 356]
            print(f"{product}: Reorder Alert (Stock: {stock})")
        else: # Print Stock OK otherwise [cite: 357]
            print(f"{product}: Stock OK (Stock: {stock})")

inventory = {"Laptops": 10, "Mice": 45, "Keyboards": 12, "Monitors": 20}
check_inventory(inventory)
```

# --- Problem 3: Student Result Processing System ---

```python
def process_student_result(marks):
    # Use loop to calculate total [cite: 376]
    total = 0
    for mark in marks:
        total += mark
        
    average = total / len(marks) # Calculates average marks [cite: 370]
    
    # Return result from function [cite: 377]
    if average >= 50: # Prints Pass if average >= 50 else Fail [cite: 371]
        return f"Average: {average:.2f} - Pass"
    else:
        return f"Average: {average:.2f} - Fail"

print("\n--- Problem 3 Output ---")
student_marks = [45, 60, 55, 70, 48]
print(process_student_result(student_marks))
```

# --- Problem 4: Cab Fare Estimator with Retry Option ---

```python
def calculate_cab_fare():
    base_fare = 50 # Base fare = 50 [cite: 383]
    per_km_rate = 12 # 12 per km [cite: 384]
    
    while True: # Allow retry using while loop [cite: 386]
        try:
            distance = float(input("\nEnter distance traveled in km: "))
            if distance < 0:
                print("Distance cannot be negative. Try again.")
                continue
                
            peak_hour = input("Is it peak hour? (yes/no): ").strip().lower()
            
            fare = base_fare + (distance * per_km_rate)
            
            if peak_hour == 'yes':
                fare = fare + (fare * 0.25) # Peak hour 25% extra [cite: 385]
                
            print(f"Total Cab Fare: ₹{fare:.2f}")

            retry = input("Calculate another fare? (yes/no): ").strip().lower()
            if retry != 'yes':
                print("Thank you for using our service.")
                break
        except ValueError:
            print("Invalid input. Please enter valid numbers for distance.")

print("\n--- Problem 4 Output ---")
calculate_cab_fare() # Uncomment this line to run the interactive prompt
```

# --- Problem 5: Employee Attendance Eligibility Checker ---

```python
def check_attendance_eligibility(attendance_list):
    present_count = 0
    # Use loop to count present days [cite: 404]
    for status in attendance_list:
        if status == "P":
            present_count += 1
            
    # Calculates attendance percentage [cite: 398]
    percentage = (present_count / len(attendance_list)) * 100
    
    # Returns Eligible if >= 75% else Not Eligible [cite: 399]
    if percentage >= 75:
        return f"Attendance: {percentage:.1f}% - Eligible"
    else:
        return f"Attendance: {percentage:.1f}% - Not Eligible"

print("\n--- Problem 5 Output ---")
attendance_record = ["P", "P", "A", "P", "P", "A", "P", "P"]
print(check_attendance_eligibility(attendance_record))
```

# --- Problem 6: Password Strength Checker ---

```python
def check_password_strength(password):
    # Minimum length = 8 [cite: 412]
    if len(password) < 8:
        return "Weak: Password must be at least 8 characters long."
        
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    # Use loop over string [cite: 419]
    for char in password:
        if char.isdigit(): # Must contain at least one digit [cite: 413]
            has_digit = True
        elif char in special_chars: # Must contain at least one special character (@#$) [cite: 414]
            has_special = True
            
    # Use conditions inside function [cite: 421]
    if has_digit and has_special:
        return "Strong Password!"
    elif not has_digit:
        return "Weak: Password must contain at least one digit."
    else:
        return f"Weak: Password must contain at least one special character ({special_chars})."

print("\n--- Problem 6 Output ---")
print(check_password_strength("Innomatics@123"))
print(check_password_strength("weakpass"))
```
